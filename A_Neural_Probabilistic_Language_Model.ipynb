Model based on https://www.jmlr.org/papers/volume3/bengio03a/bengio03a.pdf

```python
import torch 
import torch.nn.functional as F 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
words = open("names.txt", "r").read().split()
```

```python
# vocap as in bigram model 
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)} 
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()} 
print(itos)
```

```python
# build up the data set of sequences of characters
block_size = 3 # how many characters to predict the next 
X, Y = [], []

for word in words:
    context = [0] * block_size # initaite a list of zeros == mapped to > '.' for the sequence of characters
#     print(word)
    for ch in word + '.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
#         print([itos[con] for con in context])
#         print(''.join(itos[i] for i in context), '-->',itos[ix] )
        context = context[1:] + [ix] # fill up the list with the next char index 
X = torch.tensor(X)
Y = torch.tensor(Y)
```

## Implementation 

```python
g = torch.Generator().manual_seed(2147483647) # for reproducability
C = torch.randn((27,2), generator=g) # 2D embedding for each of the 27 characters
W1 = torch.randn((6,200), generator=g)
b1 = torch.randn(200, generator=g)
W2 = torch.randn((200, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
for p in parameters:
    p.requires_grad = True
sum(p.nelement() for p in parameters )
```

```python
# rough learning rate finder 
lre = torch.linspace(-3, 0, 1000)
lrs = 10**lre
```

```python
bs = 64
# lri = []
lossi = []
stepi = []

for i in range(20000):
    # contruct the mini bacth 
    ix = torch.randint(0, Xtr.shape[0], (bs,))
    # forward
    emb = C[X[ix]]
    h = torch.tanh(emb.view(-1, 6) @ W1 + b1) # broadcasting working okay 
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Ytr[ix]) 
# counts=logits.exp(), prob = counts/ counts.sum(1, keepdims=True), loss = -prob[np.arange(32 examples), Y].log.mean()
    # backward
    
    for p in parameters:
        p.grad = None
    loss.backward() # populate the gradient 
    
#     lr = lrs[i]
    lr = 10**-2
    for p in parameters:
        p.data += - lr * p.grad
      
    stepi.append(i)
    lossi.append(loss.item())
#     lri.append(lre[i])
        
print(loss.item())
```

## Introduce Training Splits

```python
# training split, developement/validation split (used for hyperparams), test split
# 80%, 10%, 10%

# build the dataset
block_size = 3 # context length: how many characters do we take to predict the next one.

def build_dataset(words):  
  X, Y = [], []
  for w in words:

    #print(w)
    context = [0] * block_size
    for ch in w + '.':
      ix = stoi[ch]
      X.append(context)
      Y.append(ix)
      #print(''.join(itos[i] for i in context), '-->', itos[ix])
      context = context[1:] + [ix] # crop and append

  X = torch.tensor(X)
  Y = torch.tensor(Y)
  print(X.shape, Y.shape)
  return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

Xtr, Ytr = build_dataset(words[:n1])
Xdev, Ydev = build_dataset(words[n1:n2])
Xte, Yte = build_dataset(words[n2:])
```
